```python
mean_e = np.mean(Vexc_b0, axis=1)
std_dev_e = np.std(Vexc_b0, axis=1)
mean_i = np.mean(Vinh_b0, axis=1)
std_dev_i = np.std(Vinh_b0, axis=1)

mean_e60 = np.mean(Vexc_b60, axis=1)
std_dev_e60 = np.std(Vexc_b60, axis=1)
mean_i60 = np.mean(Vinh_b60, axis=1)
std_dev_i60 = np.std(Vinh_b60, axis=1)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10,5))
fig.suptitle('Excitatory neurons', y=0.95)
axes[0].hist(mean_e, bins=100, label='b0')
axes[1].hist(std_dev_e, bins=100, label='b0')
axes[0].hist(mean_e60, bins=100, label='b60')
axes[1].hist(std_dev_e60, bins=100, label='b60')
axes[0].legend()
axes[1].legend()
axes[0].set(title='Mean of voltage')
axes[1].set(title='Standard Deviation of voltage')

fig, axes = plt.subplots(1, 2, figsize=(10,5))
fig.suptitle('Inhibitory neurons', y=0.95)
axes[0].hist(mean_i, bins=100, label='b0')
axes[1].hist(std_dev_i, bins=100, label='b0')
axes[0].hist(mean_i60, bins=100, label='b60')
axes[1].hist(std_dev_i60, bins=100, label='b60')
axes[0].legend()
axes[1].legend()
axes[0].set(title='Mean of voltage')
axes[1].set(title='Standard Deviation of voltage')
plt.show()
```

```python
import scipy.stats as stats
def mann_whitney_u_test(distribution_1, distribution_2):
    """
    Perform the Mann-Whitney U Test, comparing two different distributions.
    Args:
       distribution_1: List. 
       distribution_2: List.
    Outputs:
        u_statistic: Float. U statisitic for the test.
        p_value: Float.
    """
    u_statistic, p_value = stats.mannwhitneyu(distribution_1, distribution_2)
    return u_statistic, p_value
```

```python
print(mann_whitney_u_test(mean_e, mean_e60))
print(mann_whitney_u_test(mean_i, mean_i60))
print(mann_whitney_u_test(std_dev_e, std_dev_e60))
print(mann_whitney_u_test(std_dev_i, std_dev_i60))
```
